```python
from pathlib import Path
import math
import warnings

import networkx as nx
import numpy as np
import pandas as pd
import statsmodels.api as sm

warnings.filterwarnings("ignore")

# ============================================================
# SETTINGS
# ============================================================
EPL_FOLDER = Path(r"C:\Users\JMPark\Documents\dhs_dashboard\data\threaded\epl")
OUTPUT_DIR = Path(r"C:\Users\JMPark\Documents\dhs_dashboard\outputs")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

PRINT_SAMPLE_ROWS = 3
MIN_ROWS_FOR_MODEL = 30


# ============================================================
# LOGGING HELPERS
# ============================================================
def log(msg: str):
    print(f"[INFO] {msg}")


def warn(msg: str):
    print(f"[WARN] {msg}")


def fail(msg: str):
    print(f"[ERROR] {msg}")


# ============================================================
# GENERIC HELPERS
# ============================================================
def bounded(value: float, low: float = 0.0, high: float = 1.0) -> float:
    return max(low, min(high, float(value)))


def minmax(series: pd.Series) -> pd.Series:
    s = pd.to_numeric(series, errors="coerce").astype(float)
    if len(s) == 0:
        return pd.Series(dtype=float)
    if s.nunique(dropna=True) <= 1:
        return pd.Series(np.zeros(len(s)), index=s.index)
    return (s - s.min()) / (s.max() - s.min())


def zsafe(series: pd.Series) -> pd.Series:
    s = pd.to_numeric(series, errors="coerce").astype(float)
    sd = s.std()
    mean = s.mean()
    if pd.isna(sd) or sd == 0:
        return pd.Series(np.zeros(len(s)), index=s.index)
    return (s - mean) / sd


def safe_autocorr(series: pd.Series) -> float:
    s = pd.to_numeric(series, errors="coerce").dropna()
    if len(s) < 4 or s.nunique() <= 1:
        return 0.0
    return float(np.clip(s.autocorr(lag=1), -1, 1))


# ============================================================
# FILE LOADING
# ============================================================
def list_epl_video_ids(folder_path: Path) -> list[str]:
    if not folder_path.exists():
        raise FileNotFoundError(f"Folder not found: {folder_path}")

    video_ids = []
    for f in folder_path.glob("*_comments_v2.*"):
        name = f.name
        for suffix in ["_comments_v2.csv", "_comments_v2.xlsx", "_comments_v2.xls"]:
            if name.endswith(suffix):
                video_ids.append(name.replace(suffix, ""))
                break
    return sorted(set(video_ids))


def read_table_any(path: Path) -> pd.DataFrame:
    if path.suffix.lower() == ".csv":
        for enc in ("utf-8", "utf-8-sig", "latin1", "cp1252"):
            try:
                return pd.read_csv(path, encoding=enc, on_bad_lines="skip")
            except Exception:
                continue
        raise ValueError(f"Could not read CSV: {path.name}")
    return pd.read_excel(path)


def standardize_comments(df: pd.DataFrame, video_id: str) -> pd.DataFrame:
    out = df.copy()
    out.columns = [str(c).strip() for c in out.columns]

    rename_map = {}
    for c in out.columns:
        cl = c.lower()
        if cl in {"video_id", "videoid"}:
            rename_map[c] = "video_id"
        elif cl in {"comment_id", "commentid", "node_id", "id"}:
            rename_map[c] = "node_id"
        elif cl in {"comment", "text", "comment_text", "body", "message"}:
            rename_map[c] = "text"
        elif cl in {"author", "user", "username", "author_name", "channel"}:
            rename_map[c] = "user"
        elif cl in {"time", "timestamp", "published_at", "publishedat", "date"}:
            rename_map[c] = "timestamp"

    out = out.rename(columns=rename_map)

    if "video_id" not in out.columns:
        out["video_id"] = video_id
    if "node_id" not in out.columns:
        out["node_id"] = [f"{video_id}_comment_{i}" for i in range(len(out))]
    if "text" not in out.columns:
        out["text"] = ""
    if "user" not in out.columns:
        out["user"] = "unknown_user"
    if "timestamp" not in out.columns:
        out["timestamp"] = pd.NaT

    out["parent_id"] = pd.NA
    out["type"] = "comment"

    return out[["video_id", "node_id", "parent_id", "text", "user", "timestamp", "type"]]


def standardize_replies(df: pd.DataFrame, video_id: str) -> pd.DataFrame:
    out = df.copy()
    out.columns = [str(c).strip() for c in out.columns]

    rename_map = {}
    for c in out.columns:
        cl = c.lower()
        if cl in {"video_id", "videoid"}:
            rename_map[c] = "video_id"
        elif cl in {"comment_id", "parent_id", "reply_to_comment_id", "reply_to", "parent"}:
            rename_map[c] = "parent_id"
        elif cl in {"reply_id", "node_id", "id"}:
            rename_map[c] = "node_id"
        elif cl in {"reply", "text", "reply_text", "body", "message"}:
            rename_map[c] = "text"
        elif cl in {"author", "user", "username", "author_name", "channel"}:
            rename_map[c] = "user"
        elif cl in {"time", "timestamp", "published_at", "publishedat", "date"}:
            rename_map[c] = "timestamp"

    out = out.rename(columns=rename_map)

    if "video_id" not in out.columns:
        out["video_id"] = video_id
    if "parent_id" not in out.columns:
        out["parent_id"] = pd.NA
    if "node_id" not in out.columns:
        out["node_id"] = [f"{video_id}_reply_{i}" for i in range(len(out))]
    if "text" not in out.columns:
        out["text"] = ""
    if "user" not in out.columns:
        out["user"] = "unknown_user"
    if "timestamp" not in out.columns:
        out["timestamp"] = pd.NaT

    out["type"] = "reply"

    return out[["video_id", "node_id", "parent_id", "text", "user", "timestamp", "type"]]


def load_single_epl_video(folder_path: Path, video_id: str) -> pd.DataFrame:
    comment_file = None
    for ext in [".csv", ".xlsx", ".xls"]:
        p = folder_path / f"{video_id}_comments_v2{ext}"
        if p.exists():
            comment_file = p
            break

    if comment_file is None:
        warn(f"No comment file found for {video_id}")
        return pd.DataFrame()

    df_comments = standardize_comments(read_table_any(comment_file), video_id)

    reply_file = None
    for ext in [".csv", ".xlsx", ".xls"]:
        p = folder_path / f"{video_id}_replies_v2{ext}"
        if p.exists():
            reply_file = p
            break

    if reply_file is not None:
        df_replies = standardize_replies(read_table_any(reply_file), video_id)
        out = pd.concat([df_comments, df_replies], ignore_index=True)
    else:
        warn(f"No reply file found for {video_id}; using comments only")
        out = df_comments.copy()

    out["timestamp"] = pd.to_datetime(out["timestamp"], errors="coerce")
    out = out.dropna(subset=["node_id", "user"]).reset_index(drop=True)
    return out


def compute_reply_depth(df: pd.DataFrame) -> pd.Series:
    parent_lookup = dict(zip(df["node_id"], df["parent_id"]))
    memo = {}

    def depth(node_id):
        if node_id in memo:
            return memo[node_id]
        parent = parent_lookup.get(node_id)
        if pd.isna(parent) or parent is None:
            memo[node_id] = 0
            return 0
        if parent == node_id:
            memo[node_id] = 1
            return 1
        memo[node_id] = 1 + depth(parent)
        return memo[node_id]

    return df["node_id"].map(depth).fillna(0).astype(int)


# ============================================================
# CONVERSION / PREP
# ============================================================
def convert_epl_to_dhs(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()

    required = {"video_id", "node_id", "parent_id", "user", "timestamp", "type"}
    missing = required - set(out.columns)
    if missing:
        raise ValueError(f"Missing columns after loading: {sorted(missing)}")

    out["post_id"] = out["node_id"].astype(str)
    out["reply_to_post_id"] = out["parent_id"]
    out["event_name"] = out["video_id"].astype(str)
    out["user_id"] = out["user"].astype(str)
    out["timestamp"] = pd.to_datetime(out["timestamp"], errors="coerce")
    out["reply_depth"] = compute_reply_depth(out)

    # Placeholder semantic fields for now.
    is_reply = out["type"].eq("reply")
    out["comment_sentiment"] = np.where(is_reply, 0.0, 0.0)
    out["reply_sentiment"] = np.where(is_reply, 0.0, np.nan)
    out["topic_similarity"] = 0.5
    out["topic_entropy"] = 0.5

    out["platform"] = "YouTube"
    out["source_type"] = "public"
    out["is_elite_source"] = 0
    out["cluster_id"] = "Cluster 1"

    keep_cols = [
        "timestamp",
        "event_name",
        "video_id",
        "user_id",
        "post_id",
        "reply_to_post_id",
        "reply_depth",
        "comment_sentiment",
        "reply_sentiment",
        "topic_similarity",
        "topic_entropy",
        "platform",
        "source_type",
        "is_elite_source",
        "cluster_id",
    ]
    return out[keep_cols]


def prepare_data(df: pd.DataFrame) -> pd.DataFrame:
    data = df.copy()
    data["timestamp"] = pd.to_datetime(data["timestamp"], errors="coerce")

    for col in [
        "reply_depth",
        "comment_sentiment",
        "reply_sentiment",
        "topic_similarity",
        "topic_entropy",
    ]:
        data[col] = pd.to_numeric(data[col], errors="coerce")

    if "platform" not in data.columns:
        data["platform"] = "Unknown"
    if "source_type" not in data.columns:
        data["source_type"] = "public"
    if "is_elite_source" not in data.columns:
        data["is_elite_source"] = 0
    if "cluster_id" not in data.columns:
        data["cluster_id"] = "Cluster 1"

    data = data.dropna(subset=["timestamp", "event_name", "user_id", "post_id"]).copy()
    data["joint_valence_gap"] = (data["reply_sentiment"] - data["comment_sentiment"]).abs()
    data["is_reply"] = data["reply_to_post_id"].notna().astype(int)
    data["hour"] = data["timestamp"].dt.floor("h")
    return data


# ============================================================
# NETWORK / DHS
# ============================================================
def build_network(df: pd.DataFrame) -> nx.Graph:
    g = nx.Graph()
    for _, row in df.iterrows():
        user = str(row["user_id"])
        g.add_node(user, source_type=row.get("source_type", "public"))

    parent_map = df.set_index("post_id")["user_id"].to_dict()
    reply_rows = df[df["reply_to_post_id"].notna()]
    for _, row in reply_rows.iterrows():
        src = str(row["user_id"])
        tgt = str(parent_map.get(row["reply_to_post_id"], f"post::{row['reply_to_post_id']}"))
        if src == tgt:
            continue
        if g.has_edge(src, tgt):
            g[src][tgt]["weight"] += 1
        else:
            g.add_edge(src, tgt, weight=1)
    return g


def compute_network_scores(g: nx.Graph):
    if g.number_of_nodes() == 0:
        return 0.0, 0.0, 0.0, {}, {}, {}

    degree = nx.degree_centrality(g)
    betweenness = nx.betweenness_centrality(g, normalized=True)

    try:
        eigenvector = nx.eigenvector_centrality_numpy(g)
    except Exception:
        eigenvector = {n: 0.0 for n in g.nodes}

    centralization = float(np.mean(list(eigenvector.values()))) if eigenvector else 0.0
    bridging = float(np.mean(list(betweenness.values()))) if betweenness else 0.0

    try:
        communities = nx.community.louvain_communities(g, seed=42)
        modularity = nx.community.modularity(g, communities)
    except Exception:
        modularity = 0.0

    return bounded(centralization * 4), bounded(bridging * 8), bounded(modularity), degree, betweenness, eigenvector


def compute_hourly(df: pd.DataFrame) -> pd.DataFrame:
    hourly = (
        df.groupby(["video_id", "hour"])
        .agg(
            comments=("post_id", "count"),
            replies=("is_reply", "sum"),
            joint_valence_gap=("joint_valence_gap", "mean"),
            topic_entropy=("topic_entropy", "mean"),
            topic_similarity=("topic_similarity", "mean"),
            avg_reply_depth=("reply_depth", "mean"),
            elite_share=("is_elite_source", "mean"),
        )
        .reset_index()
    )
    hourly["topic_fragmentation"] = 1 - hourly["topic_similarity"].fillna(hourly["topic_similarity"].median())
    return hourly


def compute_dhs_components_for_video(df_video: pd.DataFrame) -> pd.DataFrame:
    if df_video.empty:
        return pd.DataFrame()

    g = build_network(df_video)
    centralization, bridging, modularity, degree, betweenness, eigenvector = compute_network_scores(g)
    hourly = compute_hourly(df_video)
    hourly = hourly.sort_values("hour").reset_index(drop=True)

    avg_reply_depth = float(df_video.loc[df_video["is_reply"] == 1, "reply_depth"].mean(skipna=True) or 0.0)
    hourly["perspective_instability"] = hourly["joint_valence_gap"].expanding().std().fillna(0.0)
    hourly["participation_volatility"] = hourly["replies"].expanding().std().fillna(0.0)

    # use full-series autocorr as a video-level placeholder repeated across hours
    lag1 = safe_autocorr(hourly["replies"])
    hourly["lag1_autocorr"] = lag1

    joint_div_n = bounded(float(hourly["joint_valence_gap"].mean(skipna=True)) / 1.2)
    topic_frag_n = bounded(float(hourly["topic_fragmentation"].mean(skipna=True)))
    reply_depth_n = bounded(avg_reply_depth / 6.0)

    hourly["perspective_instability_n"] = bounded(float(hourly["perspective_instability"].mean(skipna=True)) / 0.5)
    hourly["participation_volatility_n"] = hourly["participation_volatility"] / max(20.0, hourly["replies"].mean() + 1)
    hourly["participation_volatility_n"] = hourly["participation_volatility_n"].clip(0, 1)
    hourly["lag1_n"] = bounded((lag1 + 1) / 2)

    hourly["perspective_divergence"] = bounded(0.55 * joint_div_n + 0.45 * topic_frag_n)
    hourly["interaction_structure"] = bounded(
        0.45 * centralization + 0.25 * (1 - bridging) + 0.20 * modularity + 0.10 * reply_depth_n
    )
    hourly["temporal_instability"] = (
        0.40 * hourly["perspective_instability_n"]
        + 0.35 * hourly["participation_volatility_n"]
        + 0.25 * hourly["lag1_n"]
    ).clip(0, 1)

    hourly["risk_pressure"] = (
        0.38 * hourly["perspective_divergence"]
        + 0.32 * hourly["interaction_structure"]
        + 0.30 * hourly["temporal_instability"]
    ).clip(0, 1)
    hourly["rolling_dhs"] = 1 - hourly["risk_pressure"]

    hourly["centralization"] = centralization
    hourly["bridging"] = bridging
    hourly["modularity"] = modularity
    hourly["avg_reply_depth_video"] = avg_reply_depth
    return hourly


# ============================================================
# AMI
# ============================================================
def compute_cross_cluster_hourly(df: pd.DataFrame) -> pd.DataFrame:
    tmp = df.copy()
    parent_cluster = tmp.set_index("post_id")["cluster_id"].to_dict()
    tmp["parent_cluster"] = tmp["reply_to_post_id"].map(parent_cluster)
    tmp["cross_cluster"] = (
        tmp["reply_to_post_id"].notna()
        & tmp["cluster_id"].notna()
        & tmp["parent_cluster"].notna()
        & (tmp["cluster_id"] != tmp["parent_cluster"])
    ).astype(int)
    return tmp.groupby(["video_id", "hour"], as_index=False)["cross_cluster"].mean()


def compute_new_user_hourly(df: pd.DataFrame) -> pd.DataFrame:
    tmp = df.copy()
    first_seen = tmp.groupby(["video_id", "user_id"])["hour"].min().rename("first_hour").reset_index()
    tmp = tmp.merge(first_seen, on=["video_id", "user_id"], how="left")
    tmp["new_user_entry"] = (tmp["hour"] == tmp["first_hour"]).astype(int)
    return tmp.groupby(["video_id", "hour"], as_index=False)["new_user_entry"].mean()


def compute_ami_hourly(all_df: pd.DataFrame, hourly_panel: pd.DataFrame) -> pd.DataFrame:
    if hourly_panel.empty:
        return hourly_panel.copy()

    cross_hourly = compute_cross_cluster_hourly(all_df)
    entry_hourly = compute_new_user_hourly(all_df)

    out_parts = []
    for video_id, temp in hourly_panel.groupby("video_id"):
        temp = temp.sort_values("hour").reset_index(drop=True).copy()

        temp["replies_lag1"] = temp["replies"].shift(1)
        temp["comments_lag1"] = temp["comments"].shift(1)
        temp["jv_lag1"] = temp["joint_valence_gap"].shift(1)
        temp["frag_lag1"] = temp["topic_fragmentation"].shift(1)
        temp["elite_lag1"] = temp["elite_share"].shift(1)

        temp["reply_velocity"] = temp["replies"].diff()
        temp["reply_acceleration"] = temp["reply_velocity"].diff()

        temp["expected_replies"] = (
            0.55 * temp["replies_lag1"].fillna(temp["replies"].median())
            + 0.20 * temp["comments_lag1"].fillna(temp["comments"].median())
            + 0.10 * temp["elite_lag1"].fillna(0) * temp["comments"].median()
            + 0.15 * (1 - temp["frag_lag1"].fillna(temp["topic_fragmentation"].median())) * temp["replies"].median()
        )

        temp["excess_attention"] = temp["replies"] - temp["expected_replies"]
        out_parts.append(temp)

    temp_all = pd.concat(out_parts, ignore_index=True)
    temp_all = temp_all.merge(cross_hourly, on=["video_id", "hour"], how="left")
    temp_all = temp_all.merge(entry_hourly, on=["video_id", "hour"], how="left")

    temp_all["cross_cluster"] = temp_all["cross_cluster"].fillna(0.0)
    temp_all["new_user_entry"] = temp_all["new_user_entry"].fillna(0.0)
    temp_all["reply_acceleration"] = temp_all["reply_acceleration"].fillna(0.0)

    for video_id, idx in temp_all.groupby("video_id").groups.items():
        temp_all.loc[idx, "z_excess_attention"] = zsafe(temp_all.loc[idx, "excess_attention"])
        temp_all.loc[idx, "z_reply_acceleration"] = zsafe(temp_all.loc[idx, "reply_acceleration"])
        temp_all.loc[idx, "z_cross_cluster"] = zsafe(temp_all.loc[idx, "cross_cluster"])
        temp_all.loc[idx, "z_new_user_entry"] = zsafe(temp_all.loc[idx, "new_user_entry"])

    temp_all["ami_score"] = (
        0.35 * temp_all["z_excess_attention"].fillna(0)
        + 0.25 * temp_all["z_reply_acceleration"].fillna(0)
        + 0.20 * temp_all["z_cross_cluster"].fillna(0)
        + 0.20 * temp_all["z_new_user_entry"].fillna(0)
    )

    temp_all["ami_spike"] = 0
    for video_id, idx in temp_all.groupby("video_id").groups.items():
        scores = temp_all.loc[idx, "ami_score"]
        cutoff = scores.quantile(0.90) if len(scores) >= 10 else scores.max()
        temp_all.loc[idx, "ami_spike"] = (scores >= cutoff).astype(int)

    return temp_all


# ============================================================
# MODELING
# ============================================================
def run_pooled_causal_test(panel: pd.DataFrame, horizon: int = 1) -> pd.DataFrame:
    if panel.empty:
        return pd.DataFrame()

    temp = panel.sort_values(["video_id", "hour"]).copy()
    temp["future_dhs"] = temp.groupby("video_id")["rolling_dhs"].shift(-horizon)
    temp["ami_x_bridging"] = temp["ami_score"] * (1 - temp["bridging"])
    temp["ami_x_fragmentation"] = temp["ami_score"] * temp["topic_fragmentation"].fillna(0)

    needed = [
        "future_dhs",
        "ami_score",
        "rolling_dhs",
        "bridging",
        "topic_fragmentation",
        "ami_x_bridging",
        "ami_x_fragmentation",
    ]
    model_df = temp[needed].dropna()

    log(f"Rows available for pooled model: {len(model_df)}")
    if len(model_df) < MIN_ROWS_FOR_MODEL:
        warn("Not enough rows for pooled model")
        return pd.DataFrame()

    X = model_df[[
        "ami_score",
        "rolling_dhs",
        "bridging",
        "topic_fragmentation",
        "ami_x_bridging",
        "ami_x_fragmentation",
    ]]
    X = sm.add_constant(X)
    y = model_df["future_dhs"]

    model = sm.OLS(y, X).fit()
    conf = model.conf_int()

    out = pd.DataFrame({
        "term": model.params.index,
        "coef": model.params.values,
        "std_err": model.bse.values,
        "t": model.tvalues.values,
        "p_value": model.pvalues.values,
        "ci_low": conf[0].values,
        "ci_high": conf[1].values,
    })
    return out


# ============================================================
# MAIN PIPELINE
# ============================================================
def main():
    log("Starting all-video EPL analysis")
    log(f"Input folder: {EPL_FOLDER}")
    log(f"Output folder: {OUTPUT_DIR}")

    video_ids = list_epl_video_ids(EPL_FOLDER)
    log(f"Found {len(video_ids)} videos")
    if not video_ids:
        raise RuntimeError("No EPL videos found")

    all_video_frames = []
    issue_log = []

    for i, video_id in enumerate(video_ids, start=1):
        try:
            log(f"[{i}/{len(video_ids)}] Loading {video_id}")
            raw = load_single_epl_video(EPL_FOLDER, video_id)
            if raw.empty:
                warn(f"{video_id}: empty raw file after loading")
                issue_log.append({"video_id": video_id, "issue": "empty raw"})
                continue

            converted = convert_epl_to_dhs(raw)
            prepared = prepare_data(converted)
            if prepared.empty:
                warn(f"{video_id}: empty after prepare_data")
                issue_log.append({"video_id": video_id, "issue": "empty prepared"})
                continue

            all_video_frames.append(prepared)
            log(f"{video_id}: rows={len(prepared):,}, users={prepared['user_id'].nunique():,}")

        except Exception as e:
            fail(f"{video_id}: {e}")
            issue_log.append({"video_id": video_id, "issue": str(e)})

    if not all_video_frames:
        raise RuntimeError("No video data could be processed")

    full_df = pd.concat(all_video_frames, ignore_index=True)
    log(f"Combined row count: {len(full_df):,}")
    log(f"Unique videos processed: {full_df['video_id'].nunique()}")

    log("Computing hourly DHS panel video by video")
    panel_parts = []
    for video_id, df_video in full_df.groupby("video_id"):
        try:
            hourly_video = compute_dhs_components_for_video(df_video)
            if not hourly_video.empty:
                panel_parts.append(hourly_video)
        except Exception as e:
            fail(f"DHS failed for {video_id}: {e}")
            issue_log.append({"video_id": video_id, "issue": f"dhs failed: {e}"})

    if not panel_parts:
        raise RuntimeError("No hourly panel could be built")

    hourly_panel = pd.concat(panel_parts, ignore_index=True)
    log(f"Hourly panel rows: {len(hourly_panel):,}")
    log(f"Hourly panel sample:")
    print(hourly_panel.head(PRINT_SAMPLE_ROWS))

    log("Computing AMI for all videos")
    hourly_panel = compute_ami_hourly(full_df, hourly_panel)
    log("AMI sample:")
    print(hourly_panel[[
        "video_id", "hour", "rolling_dhs", "ami_score", "excess_attention",
        "reply_acceleration", "cross_cluster", "new_user_entry", "ami_spike"
    ]].head(PRINT_SAMPLE_ROWS))

    log("Running pooled causal test")
    causal_table = run_pooled_causal_test(hourly_panel, horizon=1)
    if not causal_table.empty:
        print(causal_table)
    else:
        warn("Causal table is empty")

    panel_path = OUTPUT_DIR / "epl_hourly_panel.csv"
    issues_path = OUTPUT_DIR / "epl_processing_issues.csv"
    causal_path = OUTPUT_DIR / "epl_causal_results.csv"

    hourly_panel.to_csv(panel_path, index=False)
    log(f"Saved panel: {panel_path}")

    pd.DataFrame(issue_log).to_csv(issues_path, index=False)
    log(f"Saved issues log: {issues_path}")

    if not causal_table.empty:
        causal_table.to_csv(causal_path, index=False)
        log(f"Saved causal results: {causal_path}")

    log("Done")


if __name__ == "__main__":
    main()
```
