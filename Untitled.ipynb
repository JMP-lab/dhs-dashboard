```python
from pathlib import Path
import numpy as np
import pandas as pd
import networkx as nx

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.model_selection import GroupShuffleSplit


# =========================================================
# 1. FILE READING
# =========================================================

def read_table_any(path: Path) -> pd.DataFrame:
    """
    Read CSV or Excel robustly.
    """
    if path.suffix.lower() == ".csv":
        for enc in ("utf-8", "utf-8-sig", "latin1", "cp1252"):
            try:
                return pd.read_csv(path, encoding=enc, on_bad_lines="skip")
            except Exception:
                pass
        raise ValueError(f"Could not read CSV: {path}")
    elif path.suffix.lower() in {".xlsx", ".xls"}:
        return pd.read_excel(path)
    else:
        raise ValueError(f"Unsupported file type: {path}")


# =========================================================
# 2. STANDARDIZE COMMENTS / REPLIES
# =========================================================

def standardize_comments(df: pd.DataFrame, video_id: str) -> pd.DataFrame:
    out = df.copy()
    out.columns = [str(c).strip() for c in out.columns]

    rename_map = {}
    for c in out.columns:
        cl = c.lower()
        if cl in {"video_id", "videoid"}:
            rename_map[c] = "video_id"
        elif cl in {"comment_id", "commentid", "node_id", "id"}:
            rename_map[c] = "node_id"
        elif cl in {"comment", "text", "comment_text", "body", "message"}:
            rename_map[c] = "text"
        elif cl in {"author", "user", "username", "author_name", "channel"}:
            rename_map[c] = "user"
        elif cl in {"time", "timestamp", "published_at", "publishedat", "date"}:
            rename_map[c] = "timestamp"

    out = out.rename(columns=rename_map)

    if "video_id" not in out.columns:
        out["video_id"] = video_id
    if "node_id" not in out.columns:
        out["node_id"] = [f"{video_id}_comment_{i}" for i in range(len(out))]
    if "text" not in out.columns:
        out["text"] = ""
    if "user" not in out.columns:
        out["user"] = "unknown_user"
    if "timestamp" not in out.columns:
        out["timestamp"] = pd.NaT

    out["parent_id"] = pd.NA
    out["type"] = "comment"

    return out[["video_id", "node_id", "parent_id", "text", "user", "timestamp", "type"]]


def standardize_replies(df: pd.DataFrame, video_id: str) -> pd.DataFrame:
    out = df.copy()
    out.columns = [str(c).strip() for c in out.columns]

    rename_map = {}
    for c in out.columns:
        cl = c.lower()
        if cl in {"video_id", "videoid"}:
            rename_map[c] = "video_id"
        elif cl in {"comment_id", "parent_id", "reply_to_comment_id", "reply_to", "parent"}:
            rename_map[c] = "parent_id"
        elif cl in {"reply_id", "node_id", "id"}:
            rename_map[c] = "node_id"
        elif cl in {"reply", "text", "reply_text", "body", "message"}:
            rename_map[c] = "text"
        elif cl in {"author", "user", "username", "author_name", "channel"}:
            rename_map[c] = "user"
        elif cl in {"time", "timestamp", "published_at", "publishedat", "date"}:
            rename_map[c] = "timestamp"

    out = out.rename(columns=rename_map)

    if "video_id" not in out.columns:
        out["video_id"] = video_id
    if "parent_id" not in out.columns:
        out["parent_id"] = pd.NA
    if "node_id" not in out.columns:
        out["node_id"] = [f"{video_id}_reply_{i}" for i in range(len(out))]
    if "text" not in out.columns:
        out["text"] = ""
    if "user" not in out.columns:
        out["user"] = "unknown_user"
    if "timestamp" not in out.columns:
        out["timestamp"] = pd.NaT

    out["type"] = "reply"

    return out[["video_id", "node_id", "parent_id", "text", "user", "timestamp", "type"]]


# =========================================================
# 3. LOAD ONE VIDEO
# =========================================================

def load_single_epl_video(folder_path: str, video_id: str) -> pd.DataFrame:
    folder = Path(folder_path)

    comment_file = None
    for ext in [".csv", ".xlsx", ".xls"]:
        p = folder / f"{video_id}_comments_v2{ext}"
        if p.exists():
            comment_file = p
            break

    if comment_file is None:
        raise FileNotFoundError(f"No comment file found for {video_id}")

    reply_file = None
    for ext in [".csv", ".xlsx", ".xls"]:
        p = folder / f"{video_id}_replies_v2{ext}"
        if p.exists():
            reply_file = p
            break

    comments = standardize_comments(read_table_any(comment_file), video_id)

    if reply_file is not None:
        replies = standardize_replies(read_table_any(reply_file), video_id)
        out = pd.concat([comments, replies], ignore_index=True)
    else:
        out = comments.copy()

    out["timestamp"] = pd.to_datetime(out["timestamp"], errors="coerce")
    out = out.dropna(subset=["node_id", "user", "timestamp"]).reset_index(drop=True)

    return out


# =========================================================
# 4. LOAD ALL VIDEOS
# =========================================================

def list_video_ids(folder_path: str):
    folder = Path(folder_path)
    vids = []

    for f in folder.glob("*_comments_v2.*"):
        name = f.name
        for suffix in ["_comments_v2.csv", "_comments_v2.xlsx", "_comments_v2.xls"]:
            if name.endswith(suffix):
                vids.append(name.replace(suffix, ""))
                break

    return sorted(set(vids))


def load_all_videos(folder_path: str) -> pd.DataFrame:
    video_ids = list_video_ids(folder_path)

    all_data = []
    skipped = []

    for vid in video_ids:
        try:
            df = load_single_epl_video(folder_path, vid)
            if not df.empty:
                all_data.append(df)
        except Exception as e:
            skipped.append((vid, str(e)))

    if skipped:
        print("\nSkipped videos:")
        for vid, err in skipped[:10]:
            print(f" - {vid}: {err}")
        if len(skipped) > 10:
            print(f"... and {len(skipped) - 10} more")

    if not all_data:
        return pd.DataFrame()

    out = pd.concat(all_data, ignore_index=True)
    return out


# =========================================================
# 5. REPLY DEPTH
# =========================================================

def compute_reply_depth(df: pd.DataFrame) -> pd.Series:
    parent_lookup = dict(zip(df["node_id"], df["parent_id"]))
    memo = {}

    def depth(node_id):
        if node_id in memo:
            return memo[node_id]

        parent = parent_lookup.get(node_id)

        if pd.isna(parent) or parent is None:
            memo[node_id] = 0
            return 0

        if parent == node_id:
            memo[node_id] = 1
            return 1

        memo[node_id] = 1 + depth(parent)
        return memo[node_id]

    return df["node_id"].map(depth).fillna(0).astype(int)


# =========================================================
# 6. DAILY PANEL
# =========================================================

def build_daily_panel(epl_df: pd.DataFrame) -> pd.DataFrame:
    df = epl_df.copy()
    df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
    df = df.dropna(subset=["timestamp"]).copy()

    df["date"] = df["timestamp"].dt.floor("D")
    df["is_reply"] = df["type"].eq("reply").astype(int)
    df["reply_depth"] = compute_reply_depth(df)

    # Basic daily activity
    daily = (
        df.groupby(["video_id", "date"])
        .agg(
            comments=("type", lambda x: (x == "comment").sum()),
            replies=("type", lambda x: (x == "reply").sum()),
            users=("user", "nunique"),
            avg_reply_depth=("reply_depth", "mean"),
        )
        .reset_index()
    )

    daily["volume_total"] = daily["comments"] + daily["replies"]
    daily["reply_ratio"] = daily["replies"] / daily["comments"].replace(0, np.nan)
    daily["reply_ratio"] = daily["reply_ratio"].replace([np.inf, -np.inf], np.nan).fillna(0)

    # Build daily user-reply network
    network_rows = []

    for (video_id, date), sub in df.groupby(["video_id", "date"]):
        g = nx.Graph()

        parent_user_map = sub.set_index("node_id")["user"].to_dict()

        for _, row in sub.iterrows():
            g.add_node(str(row["user"]))

        replies = sub[sub["parent_id"].notna()]

        for _, row in replies.iterrows():
            src = str(row["user"])
            tgt = str(parent_user_map.get(row["parent_id"], f"post::{row['parent_id']}"))
            if src == tgt:
                continue

            if g.has_edge(src, tgt):
                g[src][tgt]["weight"] += 1
            else:
                g.add_edge(src, tgt, weight=1)

        if g.number_of_nodes() == 0 or g.number_of_edges() == 0:
            bridging = 0.0
            centralization = 0.0
            modularity = 0.0
        else:
            btw = nx.betweenness_centrality(g, normalized=True)
            bridging = float(np.mean(list(btw.values()))) if btw else 0.0

            try:
                eig = nx.eigenvector_centrality_numpy(g)
                centralization = float(np.mean(list(eig.values()))) if eig else 0.0
            except Exception:
                centralization = 0.0

            try:
                communities = nx.community.louvain_communities(g, seed=42)
                modularity = float(nx.community.modularity(g, communities))
            except Exception:
                modularity = 0.0

        network_rows.append(
            {
                "video_id": video_id,
                "date": date,
                "bridging": bridging,
                "centralization": centralization,
                "modularity": modularity,
            }
        )

    network_daily = pd.DataFrame(network_rows)
    daily = daily.merge(network_daily, on=["video_id", "date"], how="left")

    # Sort for panel operations
    daily = daily.sort_values(["video_id", "date"]).reset_index(drop=True)

    # Rolling volatilities within each video
    for col in ["replies", "reply_ratio", "bridging", "centralization"]:
        daily[f"{col}_volatility"] = (
            daily.groupby("video_id")[col]
            .rolling(window=7, min_periods=3)
            .std()
            .reset_index(level=0, drop=True)
            .fillna(0)
        )

    return daily


# =========================================================
# 7. LAYER SCORES
# =========================================================

def zscore(s: pd.Series) -> pd.Series:
    s = pd.to_numeric(s, errors="coerce").fillna(0)
    sd = s.std()
    if pd.isna(sd) or sd == 0:
        return pd.Series(np.zeros(len(s)), index=s.index)
    return (s - s.mean()) / sd


def add_layer_scores(daily: pd.DataFrame) -> pd.DataFrame:
    df = daily.copy()

    # Surface activity layer
    df["activity_level"] = (
        zscore(df["volume_total"]) + zscore(df["reply_ratio"])
    ) / 2

    # Simple interaction-pressure proxy for now
    # This is not semantic perspective yet
    df["perspective_proxy"] = (
        zscore(df["avg_reply_depth"]) + zscore(df["reply_ratio"])
    ) / 2

    # Network openness layer
    df["network_level"] = (
        zscore(df["bridging"]) - zscore(df["centralization"]) - zscore(df["modularity"])
    ) / 3

    # Temporal instability
    df["temporal_instability"] = (
        zscore(df["replies_volatility"]) +
        zscore(df["reply_ratio_volatility"]) +
        zscore(df["bridging_volatility"]) +
        zscore(df["centralization_volatility"])
    ) / 4

    # Cross-layer divergence
    df["layer_divergence"] = np.sqrt(
        (
            (df["activity_level"] - df["perspective_proxy"]) ** 2 +
            (df["activity_level"] - df["network_level"]) ** 2 +
            (df["perspective_proxy"] - df["network_level"]) ** 2
        ) / 3
    )

    return df


# =========================================================
# 8. FUTURE TRANSITION LABELS
# =========================================================

def add_transition_labels(daily: pd.DataFrame, horizon: int = 1) -> pd.DataFrame:
    df = daily.copy().sort_values(["video_id", "date"]).reset_index(drop=True)

    future_cols = [
        "reply_ratio",
        "bridging",
        "centralization",
        "modularity",
        "temporal_instability",
        "activity_level",
        "network_level"
    ]

    for col in future_cols:
        df[f"{col}_lead"] = df.groupby("video_id")[col].shift(-horizon)

    # Global thresholds
    low_bridging = df["bridging_lead"].quantile(0.33)
    high_cent = df["centralization_lead"].quantile(0.67)
    high_mod = df["modularity_lead"].quantile(0.67)
    high_instab = df["temporal_instability_lead"].quantile(0.67)
    low_instab = df["temporal_instability_lead"].quantile(0.33)
    high_reply_ratio = df["reply_ratio_lead"].quantile(0.67)
    high_activity = df["activity_level_lead"].quantile(0.67)
    high_network = df["network_level_lead"].quantile(0.67)

    # 1. Fragmentation
    df["fragmentation_next"] = (
        (df["bridging_lead"] <= low_bridging) &
        (df["modularity_lead"] >= high_mod) &
        (df["temporal_instability_lead"] >= high_instab)
    ).astype(int)

    # 2. Entrenchment
    df["entrenchment_next"] = (
        (df["bridging_lead"] <= low_bridging) &
        (df["centralization_lead"] >= high_cent)
    ).astype(int)

    # 3. Rebound
    df["rebound_next"] = (
        (df["temporal_instability"] >= df["temporal_instability"].quantile(0.67)) &
        (df["temporal_instability_lead"] <= low_instab) &
        (
            (df["reply_ratio_lead"] >= high_reply_ratio) |
            (df["activity_level_lead"] >= high_activity) |
            (df["network_level_lead"] >= high_network)
        )
    ).astype(int)

    return df


# =========================================================
# 9. MODEL FITTING WITH GROUP SPLIT
# =========================================================

def fit_model_group(df: pd.DataFrame, target: str):
    model_df = df.dropna(subset=[
        "activity_level",
        "perspective_proxy",
        "network_level",
        "temporal_instability",
        "layer_divergence",
        target
    ]).copy()

    X = model_df[
        [
            "activity_level",
            "perspective_proxy",
            "network_level",
            "temporal_instability",
            "layer_divergence"
        ]
    ]
    y = model_df[target]
    groups = model_df["video_id"]

    if y.nunique() < 2:
        print(f"{target}: only one class present, skipped.")
        return None, None, None

    gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
    train_idx, test_idx = next(gss.split(X, y, groups))

    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    model = LogisticRegression(
        max_iter=2000,
        class_weight="balanced"
    )
    model.fit(X_train, y_train)

    prob = model.predict_proba(X_test)[:, 1]
    pred = (prob >= 0.5).astype(int)

    print(f"\n===== {target} =====")
    print(classification_report(y_test, pred, digits=3))

    if len(np.unique(y_test)) == 2:
        print("ROC AUC:", round(roc_auc_score(y_test, prob), 4))
    else:
        print("ROC AUC skipped: only one class in test set")

    coef_df = pd.DataFrame({
        "feature": X.columns,
        "coef": model.coef_[0]
    }).sort_values("coef", ascending=False)

    print("\nCoefficients:")
    print(coef_df)

    out_test = model_df.iloc[test_idx][["video_id", "date"]].copy()
    out_test[f"p_{target}"] = prob
    out_test[target] = y_test.values

    return model, coef_df, out_test


# =========================================================
# 10. MAIN RUN
# =========================================================

if __name__ == "__main__":
    folder_path = r"C:\Users\JMPark\Documents\dhs_dashboard\data\threaded\epl"

    # Load all EPL videos
    raw_all = load_all_videos(folder_path)

    print("Raw rows:", len(raw_all))
    print("Videos loaded:", raw_all["video_id"].nunique() if not raw_all.empty else 0)

    if raw_all.empty:
        raise ValueError("No EPL data loaded. Check folder path and file names.")

    # Build daily panel
    daily = build_daily_panel(raw_all)
    daily = add_layer_scores(daily)
    daily = add_transition_labels(daily, horizon=1)

    print("\nDaily panel shape:", daily.shape)
    print("Unique videos in panel:", daily["video_id"].nunique())
    print("\nTarget prevalence:")
    print(daily[["fragmentation_next", "rebound_next", "entrenchment_next"]].mean())

    # Fit models
    frag_model, frag_coef, frag_test = fit_model_group(daily, "fragmentation_next")
    reb_model, reb_coef, reb_test = fit_model_group(daily, "rebound_next")
    ent_model, ent_coef, ent_test = fit_model_group(daily, "entrenchment_next")

    # Save outputs
    daily.to_csv("epl_daily_panel_all_videos.csv", index=False)

    pred_frames = []
    for x in [frag_test, reb_test, ent_test]:
        if x is not None:
            pred_frames.append(x)

    if pred_frames:
        preds = pred_frames[0]
        for extra in pred_frames[1:]:
            preds = preds.merge(extra, on=["video_id", "date"], how="outer")
        preds.to_csv("epl_transition_predictions.csv", index=False)

    print("\nSaved:")
    print(" - epl_daily_panel_all_videos.csv")
    print(" - epl_transition_predictions.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

daily = pd.read_csv("epl_daily_panel_all_videos.csv")
preds = pd.read_csv("epl_transition_predictions.csv")

daily["date"] = pd.to_datetime(daily["date"])
preds["date"] = pd.to_datetime(preds["date"])

# Merge predictions back into daily panel
df = daily.merge(preds, on=["video_id", "date"], how="left")

# Choose one video
video = df["video_id"].dropna().unique()[0]
sub = df[df["video_id"] == video].sort_values("date").copy()

# Fill missing prediction columns if needed
for col in [
    "p_fragmentation_next",
    "p_entrenchment_next",
    "p_rebound_next",
    "fragmentation_next",
    "entrenchment_next",
    "rebound_next"
]:
    if col not in sub.columns:
        sub[col] = 0

sub[[
    "p_fragmentation_next",
    "p_entrenchment_next",
    "p_rebound_next",
    "fragmentation_next",
    "entrenchment_next",
    "rebound_next"
]] = sub[[
    "p_fragmentation_next",
    "p_entrenchment_next",
    "p_rebound_next",
    "fragmentation_next",
    "entrenchment_next",
    "rebound_next"
]].fillna(0)

# Event subsets
frag = sub[sub["fragmentation_next"] == 1]
ent = sub[sub["entrenchment_next"] == 1]
reb = sub[sub["rebound_next"] == 1]

# Optional coordination measure
sub["coordination"] = (
    sub["activity_level"] +
    sub["perspective_proxy"] +
    sub["network_level"]
) / 3

# ==============================
# 3-panel figure
# ==============================
fig, axs = plt.subplots(3, 1, figsize=(14, 10), sharex=True)

# --------------------------------
# Panel 1: Early warning
# --------------------------------
axs[0].plot(
    sub["date"],
    sub["temporal_instability"],
    color="black",
    linewidth=2,
    label="Instability"
)
axs[0].set_ylabel("Instability")
axs[0].set_title(f"Early Warning Signals - {video}")

ax2 = axs[0].twinx()
ax2.plot(
    sub["date"],
    sub["p_fragmentation_next"],
    color="red",
    linestyle="--",
    linewidth=2,
    label="P(Fragmentation)"
)
ax2.plot(
    sub["date"],
    sub["p_entrenchment_next"],
    color="blue",
    linestyle="--",
    linewidth=2,
    label="P(Entrenchment)"
)
ax2.plot(
    sub["date"],
    sub["p_rebound_next"],
    color="green",
    linestyle="--",
    linewidth=2,
    label="P(Rebound)"
)
ax2.set_ylabel("Probability")

lines1, labels1 = axs[0].get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
axs[0].legend(lines1 + lines2, labels1 + labels2, loc="upper right")

# --------------------------------
# Panel 2: Layer dynamics
# --------------------------------
axs[1].plot(sub["date"], sub["activity_level"], label="Activity", linewidth=2)
axs[1].plot(sub["date"], sub["network_level"], label="Network", linewidth=2)
axs[1].plot(sub["date"], sub["temporal_instability"], label="Instability", linewidth=2)
axs[1].plot(sub["date"], sub["layer_divergence"], label="Divergence", linewidth=2)

axs[1].set_ylabel("Standardized level")
axs[1].set_title("Layer Dynamics")
axs[1].legend(loc="upper right")

# --------------------------------
# Panel 3: Realized transitions
# --------------------------------
axs[2].plot(
    sub["date"],
    sub["temporal_instability"],
    color="black",
    linewidth=2,
    label="Instability"
)

axs[2].scatter(
    frag["date"],
    frag["temporal_instability"],
    color="red",
    s=90,
    label="Fragmentation"
)
axs[2].scatter(
    ent["date"],
    ent["temporal_instability"],
    color="blue",
    s=90,
    label="Entrenchment"
)
axs[2].scatter(
    reb["date"],
    reb["temporal_instability"],
    color="green",
    s=90,
    label="Rebound"
)

axs[2].set_ylabel("Instability")
axs[2].set_title("Realized Transitions")
axs[2].legend(loc="upper right")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Smooth function
def smooth(x, w=3):
    return pd.Series(x).rolling(w, min_periods=1).mean()

# Normalize instability for comparison
sub["instability_norm"] = (
    sub["temporal_instability"] - sub["temporal_instability"].mean()
) / (sub["temporal_instability"].std() + 1e-6)

# Smooth probabilities
sub["p_frag_s"] = smooth(sub["p_fragmentation_next"], 3)
sub["p_ent_s"] = smooth(sub["p_entrenchment_next"], 3)
sub["p_reb_s"] = smooth(sub["p_rebound_next"], 3)

# ==============================
# Improved early warning plot
# ==============================
fig, ax = plt.subplots(figsize=(14,6))

# Instability (normalized)
ax.plot(
    sub["date"],
    sub["instability_norm"],
    color="black",
    linewidth=2,
    label="Instability (normalized)"
)

# Probabilities (scaled up slightly for visibility)
ax.plot(
    sub["date"],
    sub["p_frag_s"] * 3,
    color="red",
    linestyle="--",
    linewidth=2,
    label="P(Fragmentation)"
)

ax.plot(
    sub["date"],
    sub["p_ent_s"] * 3,
    color="blue",
    linestyle="--",
    linewidth=2,
    label="P(Entrenchment)"
)

ax.plot(
    sub["date"],
    sub["p_reb_s"] * 3,
    color="green",
    linestyle="--",
    linewidth=2,
    label="P(Rebound)"
)

# Threshold line (early warning trigger)
ax.axhline(1.0, color="gray", linestyle=":", label="Warning Threshold")

ax.legend()
plt.title(f"Early Warning Signals (Aligned Scale) - {video}")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch
import numpy as np

fig, ax = plt.subplots(figsize=(16, 6))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis("off")

# ---------- helpers ----------
def big_box(x, y, w, h, title, subtitle_lines, face="#fbfbfb"):
    patch = FancyBboxPatch(
        (x, y), w, h,
        boxstyle="round,pad=0.012,rounding_size=0.02",
        linewidth=1.2,
        edgecolor="#555555",
        facecolor=face
    )
    ax.add_patch(patch)

    ax.text(
        x + 0.02, y + h - 0.055,
        title,
        ha="left", va="top",
        fontsize=14, fontweight="bold", color="#222222"
    )

    ax.text(
        x + 0.02, y + h - 0.11,
        "\n".join(subtitle_lines),
        ha="left", va="top",
        fontsize=10.5, color="#444444"
    )

def add_arrow(x1, y1, x2, y2):
    ax.annotate(
        "",
        xy=(x2, y2),
        xytext=(x1, y1),
        arrowprops=dict(
            arrowstyle="->",
            lw=1.8,
            color="#333333"
        )
    )

def mini_bar(x, y, w=0.08, h=0.10):
    vals = [0.35, 0.65, 0.5, 0.82]
    bw = w / 6
    gap = bw * 0.45
    for i, v in enumerate(vals):
        ax.add_patch(plt.Rectangle(
            (x + i * (bw + gap), y),
            bw, h * v,
            facecolor="#9ecae1",
            edgecolor="none"
        ))

def mini_network(cx, cy, scale=0.06):
    pts = np.array([
        [-0.45,  0.05],
        [-0.15,  0.35],
        [-0.10, -0.28],
        [ 0.18,  0.20],
        [ 0.35, -0.10],
        [ 0.00,  0.00]
    ]) * scale
    edges = [(0,1),(0,2),(1,3),(2,5),(5,3),(5,4),(3,4)]
    for i, j in edges:
        ax.plot(
            [cx + pts[i,0], cx + pts[j,0]],
            [cy + pts[i,1], cy + pts[j,1]],
            color="#999999", lw=1
        )
    ax.scatter(cx + pts[:,0], cy + pts[:,1], s=20, color="#6baed6", zorder=3)

def mini_wave(x, y, w=0.11, h=0.08):
    xs = np.linspace(0, 1, 120)
    ys = 0.35 + 0.18*np.sin(xs*3.2*np.pi) + 0.28*np.exp(-((xs-0.72)/0.12)**2)
    ax.plot(x + xs*w, y + ys*h, color="#756bb1", lw=1.8)

def mini_warning(x, y, w=0.11, h=0.08):
    xs = np.linspace(0, 1, 100)
    ys = np.piecewise(
        xs,
        [xs < 0.55, xs >= 0.55],
        [lambda z: 0.18 + 0*z, lambda z: 0.18 + (z-0.55)*1.3]
    )
    ax.plot(x + xs*w, y + ys*h, color="#e6550d", lw=2)
    ax.hlines(y + 0.62*h, x, x + w, color="#888888", lw=1, linestyles="dotted")

def mini_gauge(cx, cy, r=0.05):
    theta = np.linspace(np.pi, 2*np.pi, 100)
    ax.plot(cx + r*np.cos(theta), cy + r*np.sin(theta), color="#999999", lw=1.2)
    # colored segments
    for a1, a2, c in [
        (np.pi, 1.35*np.pi, "#d7301f"),
        (1.35*np.pi, 1.65*np.pi, "#fdae6b"),
        (1.65*np.pi, 1.85*np.pi, "#a1d99b"),
        (1.85*np.pi, 2*np.pi, "#31a354"),
    ]:
        th = np.linspace(a1, a2, 30)
        ax.plot(cx + r*np.cos(th), cy + r*np.sin(th), color=c, lw=3)
    # needle
    ang = 1.72*np.pi
    ax.plot([cx, cx + 0.8*r*np.cos(ang)], [cy, cy + 0.8*r*np.sin(ang)], color="#222222", lw=1.6)
    ax.scatter([cx], [cy], s=12, color="#222222", zorder=3)

# ---------- boxes ----------
big_box(
    0.04, 0.30, 0.22, 0.42,
    "Structural Context",
    [
        "• Initial conditions",
        "• User composition",
        "• Attention allocation"
    ],
    face="#f7fbff"
)

big_box(
    0.30, 0.30, 0.22, 0.42,
    "Interaction Pattern",
    [
        "• Bridging",
        "• Clustering",
        "• Centralization"
    ],
    face="#f8f8f8"
)

big_box(
    0.56, 0.30, 0.22, 0.42,
    "Temporal Dynamics",
    [
        "• Waves",
        "• Joint valence",
        "• Volatility"
    ],
    face="#faf8ff"
)

big_box(
    0.79, 0.48, 0.17, 0.24,
    "Early Signals",
    [
        "• Instability ↑",
        "• Acceleration",
        "• Misalignment"
    ],
    face="#fff8ef"
)

big_box(
    0.79, 0.18, 0.17, 0.20,
    "DHS State",
    [
        "• DHS-Level",
        "• DHS-Instability"
    ],
    face="#f3fbf4"
)

# ---------- mini visuals ----------
mini_bar(0.17, 0.37, w=0.06, h=0.11)
mini_network(0.46, 0.40, scale=0.10)
mini_wave(0.66, 0.36, w=0.09, h=0.14)
mini_warning(0.83, 0.53, w=0.09, h=0.12)
mini_gauge(0.88, 0.23, r=0.045)

# ---------- arrows ----------
add_arrow(0.26, 0.51, 0.30, 0.51)
add_arrow(0.52, 0.51, 0.56, 0.51)

# all three to early signals
add_arrow(0.20, 0.70, 0.82, 0.68)
add_arrow(0.46, 0.70, 0.84, 0.66)
add_arrow(0.72, 0.70, 0.86, 0.64)

# early signals to dhs state
add_arrow(0.875, 0.48, 0.875, 0.38)

# ---------- title ----------
ax.text(
    0.5, 0.92,
    "Discourse System Health (DHS) Framework",
    ha="center", va="center",
    fontsize=18, fontweight="bold", color="#222222"
)

ax.text(
    0.5, 0.875,
    "Context, interaction, and temporal dynamics generate early signals that shape measured discourse state",
    ha="center", va="center",
    fontsize=10.5, color="#666666"
)

plt.tight_layout()
plt.show()
```
